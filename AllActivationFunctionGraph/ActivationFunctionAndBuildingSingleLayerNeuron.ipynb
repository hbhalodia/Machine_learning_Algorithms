```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def relu(x):
    return (max(0,x))
```

```python
x = [float(x)/100 for x in range(-100,100)] #input
y = [relu(x1) for x1 in x]  #output
```

```python
def plot_fun(x,y,title):
    plt.figure(figsize=(5,3)).gca()
    plt.plot(x,y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(title)
```

```python
#plotting relu function
plot_fun(x,y,'Relu-Function')
```

```python
from math import exp
def sigmoid(x):
    return (exp(x)/(1+exp(x)))
```

```python
x1 = [float(x1)/100 for x1 in range(-500,500)] #input
y1 = [sigmoid(x2) for x2 in x1]  #output
```

```python
#plotting relu function
plot_fun(x1,y1,'Sigmoid-Function')
```

```python
#leaky relu
al = 0.3
def Leakyrelu(x):
    if(x>=0):
        return x
    else:
        return al*x

x= [float(x)/100 for x in range(-500,500)] #input
y = [Leakyrelu(x1) for x1 in x]  #output

#plotting leaky relu function
plot_fun(x,y,'Leaky relu Function')
```

```python
#swish function
bta = 1
def swish(x):
    return 2*x*sigmoid(bta*x)
x= [float(x)/100 for x in range(-500,500)] #input
y = [swish(x2) for x2 in x]  #output

#plotting tanh function
plot_fun(x,y,'Swish Function')
```

```python
#tanh function 
def tanh(x):
    return 2 * sigmoid(2*x) - 1

x= [float(x)/100 for x in range(-500,500)] #input
y = [tanh(x2) for x2 in x]  #output

#plotting tanh function
plot_fun(x,y,'Tanh Function')
```

```python
#bunary Function
def binary(x):
    if(x>=0):
        return 1
    else :
        return 0

x= [float(x)/100 for x in range(-500,500)] #input
y = [binary(x2) for x2 in x]  #output

#plotting Binary function
plot_fun(x,y,'Binary Function')
```

```python
#step Function 
al = 4
def step(x):
    if(x>=0):
        return al
    else :
        return 0

x= [float(x)/100 for x in range(-500,500)] #input
y = [step(x1) for x1 in x]  #output

#plotting step function
plot_fun(x,y,'Step Function')
```

```python
#Linear Function
slope=2
y_intcpt=0
def linear(x):
    return slope*x+y_intcpt

x= [float(x)/100 for x in range(-500,500)] #input
y = [linear(x1) for x1 in x]  #output

#plotting lineatr function
plot_fun(x,y,'Linear Function')
```
